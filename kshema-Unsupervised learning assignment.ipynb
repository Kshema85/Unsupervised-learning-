```python
# KMEANS CLUSTERING
```

```python
wcss = []
for i in range(1, 13):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(wine)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 13), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
 kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
y_means=kmeans.fit_predict(wine)
```

```python
#  DBSCAN
```

```python
from sklearn.neighbors import NearestNeighbors # importing the library
neighb = NearestNeighbors(n_neighbors=2)
nbrs=neighb.fit(wine)
distances,indices=nbrs.kneighbors(wine) # finding the nearest neighbours
```

```python
distances = np.sort(distances, axis = 0) 
distances = distances[:, 1] 
plt.rcParams['figure.figsize'] = (5,3)
plt.plot(distances)
plt.show() 
```

```python
from sklearn.cluster import DBSCAN
# cluster the data into five clusters
dbscan = DBSCAN(eps = 35, min_samples = 4).fit(wine) # fitting the model
labels = dbscan.labels_ # getting the labels
```

```python
labels
```

```python
sil_avg2=silhouette_score(wine,labels)
```
